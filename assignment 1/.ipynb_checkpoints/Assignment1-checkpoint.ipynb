### Opening the data
Before we answer the questions of Assignment 1A, we open the four provided datasets, and append the numbers in the datasets' columns, to four lists of X and Y values.

````python
X1 = []
Y1 = []
for line in open('./data1.tsv'):
    xtemp,ytemp = line.strip().split('\t')
    X1.append(float(xtemp))
    Y1.append(float(ytemp))


X2 = []
Y2 = []
for line in open ('./data2.tsv'):
    xtemp2,ytemp2 = line.strip().split('\t')
    X2.append(float(xtemp2))
    Y2.append(float(ytemp2))


    
X3 = []
Y3 = []
for line in open ('./data3.tsv'):
    xtemp3,ytemp3 = line.strip().split('\t')
    X3.append(float(xtemp3))
    Y3.append(float(ytemp3))


X4 = []
Y4 = []
for line in open ('./data4.tsv'):
    xtemp4,ytemp4 = line.strip().split('\t')
    X4.append(float(xtemp4))
    Y4.append(float(ytemp4))

````

### Printing mean values with two decimals
-Using the numpy function mean, calculate the mean of both  x-values and  y-values for each dataset.

-Use python string formatting to print precisely two decimal places of these results to the output cell.

````python
import numpy as np

data1 = np.array([X1, Y1])

data2 = np.array([X2, Y2])

data3 = np.array([X3, Y3])

data4 = np.array([X4, Y4])

## Print the mean of both the X and Y colum with 2 decimals, for all the datasets
print "The mean values of dataset1's X and Y columns, respectively, are:"
for i in data1.mean(axis=1): 
    print "{:10.2f}".format(i)
print "The mean values of dataset2's X and Y columns, respectively, are:"
for i in data2.mean(axis=1):
    print "{:10.2f}".format(i)
print "The mean values of dataset3's X and Y columns, respectively, are:"
for i in data3.mean(axis=1):
    print "{:10.2f}".format(i)
print "The mean values of dataset4's X and Y columns, respectively, are:"
for i in data4.mean(axis=1):
    print "{:10.2f}".format(i)
    
````

## Assignment 1b asnwer
### Opening the data for assignment 1B
The following section of code opens the data needed to anaswer assignment 1B. 
A function is used to read the csv file containing the data. The data is stored in a numpy array, which is converted into a pandas dataframe.

````python
##Open data from SF OpenData
import numpy as np
import csv


def openfile(filename):
    category = []
    date = []
    time = []
    pdistrict = []
    xcoords = []
    ycoords = []
    with open(filename, 'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        reader.next()
        for row in reader:
            category += [row[1]]
            date += [row[4]]
            time += [row[5]]
            pdistrict += [row[6]]
            xcoords += [float(row[9])]
            ycoords += [float(row[10])]            
    return np.array([category, date, time, pdistrict, xcoords, ycoords]).T

    
crimedata = openfile("../SFPD_Incidents_-_from_1_January_2003.csv")

import pandas as pd

df = pd.DataFrame(crimedata)
````

## KNN Classifier
-Next, it's time to set up your model based on the actual data. You can use the code supplied in the book or try out scikit-learn's KNeighborsClassifier.

We use scikit-learn's classifier to create our model.
We create a training set, countaing 80 pct. of all the data on the focus crimes. Furthermore, we create a test set that contains the last 20 pct. of the data. We fit the model using the training data, and predict the crime category for the test data.

````python
import sys
import numpy as np
import pylab as pl


# convert string coordinates in pandas dateframes to floats


df_p[4] = df_p[4].astype(float)
df_p[5] = df_p[5].astype(float)
df_duti[4] = df_duti[4].astype(float)
df_duti[5] = df_duti[5].astype(float)
df_dn[4] = df_dn[4].astype(float)
df_dn[5] = df_dn[5].astype(float)

####Laver test og training data
from sklearn.model_selection import train_test_split

train_p, test_p = train_test_split(df_p, test_size = 0.2, random_state=42)

train_duti, test_duti = train_test_split(df_duti, test_size = 0.2, random_state=42)

train_dn, test_dn = train_test_split(df_dn, test_size = 0.2, random_state=42)

train_focus = pd.concat([train_p, train_duti, train_dn], ignore_index=True)
test_focus = pd.concat([test_p, test_duti, test_dn], ignore_index=True)

X_train = np.array([train_focus[4], train_focus[5]]).T
Y_train = np.array(train_focus[0])

X_test = np.array([test_focus[4], test_focus[5]]).T
Y_test = np.array(test_focus[0])


from sklearn.neighbors import KNeighborsClassifier
#Predicting with 5 neighbors
neigh_5 = KNeighborsClassifier(n_neighbors=5)
neigh_5.fit(X_train, Y_train) 
#Predicting with 10 neighbors
neigh_10 = KNeighborsClassifier(n_neighbors=10)
neigh_10.fit(X_train, Y_train)
#Predicting with 30 neighbors
neigh_30 = KNeighborsClassifier(n_neighbors=30)
neigh_30.fit(X_train, Y_train) 

X_5_predict = np.array(neigh_5.predict(X_test))
X_10_predict = np.array(neigh_10.predict(X_test))
X_30_predict = np.array(neigh_30.predict(X_test))


import pandas as pd

df_X_5_test = pd.DataFrame(X_test)
df_X_5_test[2] = X_5_predict

df_X_10_test = pd.DataFrame(X_test)
df_X_10_test[2] = X_10_predict

df_X_30_test = pd.DataFrame(X_test)
df_X_30_test[2] = X_30_predict
````

# Assignment 1D: Multiple Regression

Let's improve the model by incorporating the total amount of crime in the area we're considering. 

In the book, the more general model with $k$ input variables is described as

$$y_i = \alpha + \beta_1x_{i1} + \ldots + \beta_k x_{ik} + \epsilon_i.$$

We will be looking at the case when $k=2$, so our model is simply

$$y_i = \alpha + \beta_1x_{i1} + \beta_2 x_{i2} + \epsilon_i.$$

In our example, we want to include *both* the time of day (as above) and the total amount of crime in our model. This means that the $(x_{i1}, x_{i2}, y_{i})$ variables in the model has the following correspondence:
* $x_{i1}$ corresponds to time-of-day (input variable)
* $x_{i2}$ corresponds to total amount of crime in the area (input variable)
* $y_i$ corresponds to the number of `LARCENY/THEFT` crimes in that bin (output variable)

As above, one observation is a $(x_{i1}, x_{i2}, y_{i})$ and you should have 12 for each day in the dataset.

In order to make the model useful, you have transform your input variables a bit before fitting (as described in detail in the Week 4 exercises). Specifically,

* Firstly, instead of using total crime vs. the crime we're considering, remember to use differences between _average_ total crime and observed crime.
* Secondly, remember to rescale your input variables.

Explain in your own words the reasoning between the two bullets above.

Finally, create and fit the data set using multiple regression. Report $\alpha, \beta_1, \beta_2$. 

* We want to use the difference between observed crime and average crime (the deviation), since this gives us an estimate of how the given day and bin performs against the trend.
* We normalize input variables to the same units in order to adjust for a possible bias in one variable.
* We find that $\alpha = 4.26301369863$, $\beta_1 = 2.40177095$ and $\beta_2 = 0.64727773$

````python
df_time = df[df[7].isin(['05', '06','07','08','09','10','11','12','13','14','15','16'])]

x2_time = list(df_time[7])
x2_cnt = Counter(x2_time)

X2 = [i / 365 for i in x2_cnt.values()]

from sklearn import preprocessing

normalized_X1 = preprocessing.scale(X1)
normalized_X2 = preprocessing.scale(X2)

norm_data = np.array([normalized_X1, normalized_X2]).T

#regdata_lar_multiple = slope, intercept, r_value, p_value, std_err = stats.linregress(X1, Y1)
from sklearn import linear_model
multi_reg = linear_model.LinearRegression()
multi_reg.fit(norm_data, Y1)
print multi_reg.intercept_
print multi_reg.coef_
````
